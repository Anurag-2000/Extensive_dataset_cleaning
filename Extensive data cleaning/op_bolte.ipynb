```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data_cleaning_challenge.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.drop('Unnamed: 9',axis=1,inplace=True)
df.drop('Unnamed: 10',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df = df[df["Row Type"].notna()]
df.head()
```

```python
colunm_val = []
counter = 0

for i in df['Row Type']:
    if "first name" in i:
        counter+=1
    colunm_val.append(counter)

```

```python
df["iterators"] = colunm_val
df
```

```python
df_drop_rows = df[df["Row Type"]!="Row Type"]
# df_drop_rows.head(15)
```

```python
df_names = df_drop_rows[df_drop_rows["Row Type"].str.contains("first name")]
df_names
```

```python
df_names['Row Type'] = df_names['Row Type'].str.replace('first name: ','')
df_names['Iter Number'] = df_names['Iter Number'].str.replace('last name: ','')
df_names['Power1'] = df_names['Power1'].str.replace('date: ','')
df_names
```

```python
df_2= df_names
df_2.drop(columns=['Speed1','Speed2','Electricity','Effort','Weight','Torque'],inplace=True)
```

```python
df_2
```

```python
df_nonames = df_drop_rows[~df_drop_rows["Row Type"].str.contains("first name")]
df_nonames
```

```python
df_final=pd.merge(left=df_2,right=df_nonames,how='inner',on='iterators')
```

```python
df_final
```

```python
df_final.to_csv('final.csv',index=True)
```
